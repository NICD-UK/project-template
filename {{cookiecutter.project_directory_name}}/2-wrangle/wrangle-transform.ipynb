## Load Libraries

```python
import pandas
from pathlib import Path
```

## Setup

```python
user = "user-name"
#root_path = Path(f"/home/jupyter-{user}/bootcamp")
root_path = Path('../')
data_name = "income"
```

## Read Data

```python
clean_data = pandas.read_pickle(root_path / f"data/clean/{data_name}.pkl")
```

## Wrangle Data

### Filter Rows

Let's drop any rows containing missing data using the `dropna` function.

```python
wrangled_data = clean_data.dropna()
```

### Select Columns

Suppose we decide we don't want to retain the 'final_weight' and 'education_code' columns in our analysis or model building?

```python
wrangled_data = wrangled_data.drop(columns=[
    'final_weight',
    'education_code',
])
```

## Write Data

Finally, we write our wrangled data to a separate directory. Note we still have or original clean data (and our raw data) should we need to go back to it.

```python
wrangled_data.to_pickle(root_path / f"data/wrangle/{data_name}.pkl")
```
